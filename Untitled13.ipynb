```python
x = df.drop(["Price (Euro)","Weight (kg)","Product"],axis = 1)
y = df["Price (Euro)"]
```

```python
x
```

```python
categorical_cols = x.select_dtypes(include=['object']).columns

X_enc = pd.get_dummies(x, columns=categorical_cols, drop_first=True)
```

```python
X_enc = X_enc.astype(int)
X_enc
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_enc, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mod = LinearRegression()
mod.fit(X_train,y_train)
```

```python
y_pred = mod.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

```python
result = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
result.head()
```
